## AGN VARIABILITY

```python
#generate folder of files with various agn values for the same galaxy 

#create range of flux values flux: = N +- percent*N
N = 500
percent = .5
agn_vals = sgf.agn_flux(N,percent)

#generate files
for i in range(len(agn_vals)):
    J101004['agn_flux'] = agn_vals[i]
    
    output = 'J101004_agn_fluxes/agn_b16d08_psf12_sky10_agn30_var10_20' + '_agnflux_' + str(i)
    
    sgf.galaxy_gen(sys.argv, J101004, output)

    
```

```python
#get fluxes summed over an aperture 
directory = 'J101004_agn_fluxes'
background = fits.open('backgrounds/HITS/calexp_DECam_g_g_DECam_SDSS_c0001_4720_0_1520_0_410979_S15_processCcdOutputs_Blind15A_34.fits')

#real galaaxy coords 
x = 152.5179851
y = 0.3549697

#synthetic galaxy coords
sx = 662.9743900 
sy = 307.7052556 + 2048

fluxes = []

#for each file, sum over aperture radius of 1 arcsec
for file in glob.iglob(f'{directory}/*'):
    
    inject = sgf.add_galaxy(file, background, x, y)
        
    flux, fluxerr, flag = sep.sum_circle(inject, [sx], [sy], 1/.27)
    fluxes = np.append(fluxes, flux)

#print(fluxes)
```

```python
#subtract median and plot fluxes
fluxes = fluxes - np.median(fluxes)

plt.figure(figsize=(12,5))
plt.plot(fluxes,'-o')
plt.title('AGN Flux')
```
